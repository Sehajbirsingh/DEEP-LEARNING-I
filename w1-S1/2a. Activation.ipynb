Activation Function:
- The activation function transforms the weighted sum of the inputs
z (the weighted sum + bias) into the neuron's output, which introduces non-linearity. 

- Without activation functions, the network would essentially behave like a linear regression model, no matter how many layers it has.

Examples of activation functions:

**ReLU:** 
Outputs 0 if z < 0 otherwise z


**Sigmoid**:
 Outputs a value between 
0 and 1
(useful for probabilities).

**Tanh:** Outputs a value between âˆ’1 and 1
(centered around zero).

**Plot different types of activation functions**

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create a range of x

```python
# Create a range of x
x = np.arange(-10, 10)
x
```

linear activation function

```python
# linear activation function
a = 1
y = a*x
```

Plot Linear activation function

```python
plt.subplots() # open a new plot
plt.plot(x, y)
plt.title("Linear")
plt.xlabel ("x")
plt.ylabel ("F(x)")
```

Sigmoid activation function

```python
# Sigmoid activation function
y = 1 / (1 + np.exp(-x))
```

Plot Sigmoid activation function

```python
plt.subplots() # open a new plot
plt.plot(x, y)
plt.title("Sigmoid")
plt.xlabel ("x")
plt.ylabel ("F(x)")
```

Tanh activation function

```python
# Tanh activation function
y = np.tanh(x)
```

Plot Tanh activation function

```python
plt.subplots() # open a new plot
plt.plot(x, y)
plt.title("tanh")
plt.xlabel ("x")
plt.ylabel ("F(x)")
```
